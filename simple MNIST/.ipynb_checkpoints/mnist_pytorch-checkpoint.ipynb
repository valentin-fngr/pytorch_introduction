```python
import torch.nn as nn
import torchvision
import torchvision.transforms as transforms
import numpy as np 
import torch
import matplotlib.pyplot as plt
```

```python
transformation = {
    "train" : transforms.Compose([
        transforms.CenterCrop(28),
        transforms.ToTensor(),
        torchvision.transforms.Normalize((0.5,), (0.5,))
        
    ]), 
    "val" :  transforms.Compose([
        transforms.ToTensor(),
        torchvision.transforms.Normalize((0.5,), (0.5,))
    ])
}
```

```python
train_ds = torchvision.datasets.MNIST("data/", download=True, train=True,transform=transformation["train"])
val_ds = torchvision.datasets.MNIST("data/", download=True, train=False, transform=transformation["val"])
```

```python
batch_size = 4
classes = list(range(10))
```

```python
train_ds = torch.utils.data.DataLoader(
    train_ds, batch_size=batch_size, shuffle=True, num_workers=2
)

val_ds = torch.utils.data.DataLoader(
    val_ds, batch_size=batch_size, shuffle=True, num_workers=2
)

device = 'cuda' if torch.cuda.is_available() else 'cpu'
print("device :", device)
```

```python
inputs, label = iter(train_ds).next()
```

```python
print(label)
print(inputs.shape)
```

```python
# grid 

def plot_image(inp, title=None): 
    inp = inp.numpy()
    inp = inp*0.5 + 0.5
    inp = np.transpose(inp, axes=[1,2,0])
    plt.imshow(inp)
    if title is not None: 
        plt.title(title)
    plt.show()

grid = torchvision.utils.make_grid(inputs)
```

```python
plot_image(grid)
```

### simple Fully connected Net

```python
import torch.nn.functional as F

class FCNet(nn.Module): 
    
    def __init__(self, nb_classes): 
        super(FCNet, self).__init__()
        self.nb_classes = nb_classes
        self.fc1 = nn.Linear(28*28, 100)
        self.fc2 = nn.Linear(100, nb_classes)
        
    def forward(self, x): 
        x = torch.reshape(x, (-1,28*28))
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        
        return x
```

```python
from torchsummary import summary
model = FCNet(len(classes))
model.to(device)
summary(model, input_size=(1,28,28))
```

```python
optimizer = torch.optim.SGD(model.parameters(), lr=0.001, momentum=0.9)
criterion = nn.CrossEntropyLoss()
```

```python
epochs = 1

for epoch in range(epochs): 
    
    running_loss = 0.0 
    for i, data in enumerate(train_ds): 
        inputs, labels = data
        
        optimizer.zero_grad()
        
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
        if i % 2000 == 1999:    # print every 2000 mini-batches
            print(f'[{epoch + 1}, {i + 1:5d}] loss: {running_loss / 2000:.3f}')
            running_loss = 0.0

print("DONE TRAINING ! ")
        
```

```python
# save the model 
torch.save(model.state_dict(), "simple_linear_saved")
```

```python
# testing the network

images, labels = iter(val_ds).next()
grid_image = torchvision.utils.make_grid(images)
plot_image(grid_image)
```

### Good performance on this batch

```python
# let's predict
pred_out = model(images)
pred = torch.argmax(pred_out, dim=1)
plot_image(grid_image, title=f"Predicted value {pred}")
```

### Run the linear model on the entire val dataset

```python
total = 0 
correct = 0 

print("Computing validation accuracy ...")
with torch.no_grad(): 
    for i,data in enumerate(val_ds): 
        inputs, labels = data
        
        outputs = model(inputs)
        
        preds = torch.argmax(outputs, dim=1)
        total += labels.shape[0]
        correct += (preds == labels).sum().item()
        
print(f"Total accuracy : {(correct / total) * 100} % ")       
```

### Implementing A simple convnet

```python
class TinyConvNet(nn.Module): 
    
    def __init__(self): 
        super(TinyConvNet, self).__init__()
        self.conv1 = torch.nn.Conv2d(1, 125, 3)
        self.pool1 = torch.nn.MaxPool2d(2, stride=2)
        self.conv2 = torch.nn.Conv2d(125, 10, 3)
        self.pool2 = torch.nn.MaxPool2d(3, stride=2)
        self.fc1 = torch.nn.Linear(250, 32)
        self.fc2 = torch.nn.Linear(32, 10)
        
    def forward(self, x): 
        x = F.relu(self.conv1(x))
        x = self.pool1(x)
        x = F.relu(self.conv2(x))
        x = self.pool2(x)
        x = x.view(x.size(0), -1)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)

        return x
        
```

```python
model = TinyConvNet()
model.to(device)
summary(model, input_size=(1,28,28), batch_size=batch_size, device=device)
```

```python
optimizer = torch.optim.SGD(model.parameters(), lr=0.001, momentum=0.9)
criterion = nn.CrossEntropyLoss()
```

```python
# training loop 
epochs = 2 

for epoch in range(epochs): 
    total_loss = 0.0 
    
    for i, data in enumerate(train_ds): 
        inputs, labels = data
        
        optimizer.zero_grad()
        
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        total_loss += loss
        loss.backward()
        optimizer.step()
        
        if i % 2000 == 1999:    # print every 2000 mini-batches
            print(f'[{epoch + 1}, {i + 1:5d}] loss: {running_loss / 2000:.3f}')
            running_loss = 0.0

print("DONE TRAINING")
```

```python
# predictions

inputs, labels = iter(val_ds).next()
preds = model(inputs)
preds = torch.argmax(preds, dim=1)

grid_image = torchvision.utils.make_grid(inputs)
plot_image(grid_image, title=f"Preds : {preds}")
```

```python
# computing val accuracy 

total = 0 
correct = 0 

with torch.no_grad(): 
    for _, data in enumerate(val_ds): 
        
        inputs, labels = data
        preds = model(inputs)
        preds = torch.argmax(preds, dim=1)
        
        correct += (preds == labels).sum().item()
        total += labels.shape[0]
        
print(f"Validation Accuracy : {(correct / total)*100} % ")
```

```python
# save model
torch.save(model.state_dict(), "tinyconvnet_saved")
```
